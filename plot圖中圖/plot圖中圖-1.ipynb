```python
import matplotlib.pyplot as plt

fig=plt.figure()
x=[1,2,3,4,5,6,7]
y=[1,3,4,2,5,8,6]

# below are all percentage
left,bottom,width,height=0.1,0.1,0.8,0.8
ax1=fig.add_axes([left,bottom,width,height])
ax1.plot(x,y,'g',linewidth=2.0)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

ax2=fig.add_axes([0.2,0.56,0.25,0.25])
ax2.plot(y,x,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('small picture 1')


plt.axes([0.6,0.2,0.25,0.25])
plt.plot(y[::-1],x,'y')
plt.xlabel('x')
plt.ylabel('y')
plt.title('small picture 2')

plt.show()
```
