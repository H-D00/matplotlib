```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=1024
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y=np.meshgrid(x,y)


plt.contourf(X,Y,f(X,Y),9,alpha=.75,cmap=plt.cm.jet)
C=plt.contour(X,Y,f(X,Y),9,colors='black',linewidth=.3)
plt.clabel(C,inline=True,fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```
